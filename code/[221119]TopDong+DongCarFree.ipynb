## 행정동별 차없거개수 구하기 전처리

```python
df2 = pd.read_csv('차없길_법정동.csv')
df2.head()
```

```python
차없거_법정동 = pd.DataFrame(df2[['시군구명','법정동명']].value_counts())
#차없거_법정동 = 차없거_법정동.rename(columns=차없거_법정동.iloc[0])
#차없거_법정동 = 차없거_법정동.drop(차없거_법정동.index[0])
차없거_법정동.to_csv('[221119]차없거개수_법정동.csv')
```

```python
차없거개수_법정동 = pd.read_csv('[221119]차없거개수_법정동.csv')
차없거개수_법정동.head()
```

```python
차없거개수_법정동.info()
```

```python
dfDong = pd.read_excel('data/한국행정구역분류(2022.10.1.기준).xlsx', sheet_name = '행정동 및 법정동코드와의 연계표')
dfDong.rename(columns = dfDong.iloc[0], inplace= True) #첫번째행을 열이름으로
dfDong.drop(dfDong.index[0],inplace=True)#첫번째행삭제
dfDong = dfDong.astype({'행정기관코드':'int64'})
dfDong = dfDong[['시군구','법정동','행정동\n(행정기관명)']]
dfDong.head()
```

```python
행정동_차없거개수 = pd.merge(dfDong,차없거개수_법정동,how = 'left',left_on = '법정동',right_on = '법정동명' )
행정동_차없거개수 = 행정동_차없거개수[['시군구','법정동','시군구명','법정동명','행정동\n(행정기관명)','0']].dropna()
행정동_차없거개수 = 행정동_차없거개수[행정동_차없거개수['시군구']==행정동_차없거개수['시군구명']]
행정동_차없거개수.info()
```

```python
행정동_차없거개수 = 행정동_차없거개수[['시군구','행정동\n(행정기관명)','0']]
행정동_차없거개수=행정동_차없거개수.groupby(['시군구','행정동\n(행정기관명)'],as_index=False).sum()
행정동_차없거개수.columns = ['자치구', '행정동', '차없거수']
행정동_차없거개수
```

```python
행정동_차없거개수.to_csv('[221119]차없거개수_행정동.csv')
```
